```python
# For Pie chart- Population of Male and Female
tips['sex'].value_counts()
tips.sex.value_counts().plot(kind='pie')
```

```python
# For Bar chart- Population of Male and Female
tips.sex.value_counts().plot(kind='bar')
```

```python
# Total bill 
plt.hist(tips['total_bill'])
```

```python
sns.stripplot(y='tip', data=tips, jitter=True)
plt.ylabel('tips($)')
plt.show()
```

```python
#Grouping with strippplot()- Which days get more tip
sns.stripplot(x='day', y='tip', data=tips)
plt.ylabel("tip($)")
plt.show()
```

```python
sns.stripplot(x='day', y='tip', data=tips, size=4, jitter=False)
plt.ylabel("tip($)")
plt.show()
```

```python
# Whch day get more tip at a particular amount
sns.swarmplot(x='day', y='tip', data=tips, size=4)
plt.ylabel('tip($)')
plt.show()
```

```python
# Male and Female who contributing more tip
sns.swarmplot(x='day', y='tip', data=tips, hue='sex', size=4,)
plt.ylabel('tip($)')
plt.show()
```

```python
# Changine orientation
sns.swarmplot(x='tip', y='day', data=tips, hue='sex', orient='h', size=3)
plt.xlabel('tip($)')
plt.show()
```

### Box and Violin plot

```python
#plt.subplot(1,2,1)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
```

```python
#plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')

plt.show()
```

```python
plt.subplot(1,2,1)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('(tip($))')

plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data=tips)

plt.show()
```

```python
# Combining plot
sns.violinplot(x='day', y='tip', data=tips, inner=None, color='lightblue')
sns.stripplot(x='day', y='tip', data=tips, jitter=True)
plt.ylabel('tip($)')
plt.show
```

```python
# Joint plot: How much get tip on a particular bill amount
sns.jointplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
# Density Plot
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
plt.show()
```

```python
# pair plot
sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips, hue='sex')
plt.show()
```

# Titanic Data Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")

```

```python
data=pd.read_csv("/home/aman/Downloads/Data science material/train.csv")
```

```python
data
```

```python
# Data types
print(data.dtypes)
```

```python
# How many people survived by Groupby
data.groupby('Survived')['PassengerId'].count()
```

```python
f, ax= plt.subplots(figsize=(11,5))
sns.boxplot(x='Survived', y="Age", data=data)
```

```python
# Male vs Female survived
f, ax=plt.subplots(figsize=(10,8))
sns.barplot(x='Sex', y='Survived', data=data);
```

```python
f, ax=plt.subplots(figsize=(11,5))
sns.boxplot(x='Sex', y='Age', hue='Survived', data=data)
```

```python
# Who survived most(1class vs 2class vs 3class)

sns.barplot(x='Pclass', y='Survived', data=data)
```

```python
# Class with Sex

sns.barplot(x='Pclass', y='Survived', data=data, hue='Sex')
```

```python
sns.barplot(x='SibSp', y='Survived', data=data);
```

```python
sns.barplot(x='Parch', y='Survived', data=data);
```

```python
survived= data.loc[data['Survived']==1, 'Age'].dropna()
sns.distplot(survived)
plt.title('Survived')
```

```python
not_survived = data.loc[data['Survived']==0,"Age"].dropna()
sns.distplot(not_survived)
plt.title("Not Survived");
```

```python
sns.pairplot(data.dropna());
```

```python
# Pclass vs Survive
grid= sns.FacetGrid(data, col='Survived', row='Pclass', size=2.4, aspect=1.5)
grid.map(plt.hist, 'Age', alpha=.5, bins=20)
grid.add_legend()
```

```python

sns.distplot(a=data[data['Embarked']=='C']['Survived'],bins=3,kde=False)
plt.title("Cherbourg")
plt.xticks([0,1])
plt.show()
plt.title("QueensTown")
sns.distplot(a=data[data['Embarked']=='Q']['Survived'],bins=3,kde=False)
plt.xticks([0,1])

plt.show()
plt.title("Southampton")
sns.distplot(a=data[data['Embarked']=='S']['Survived'],bins=3,kde=False)
plt.xticks([0,1])

plt.show()
#Most of the Passengers embarked from Southampton
```

```python
figbi, axesbi = plt.subplots(2, 4, figsize=(16, 10))
data.groupby('Pclass')['Survived'].mean().plot(kind='barh',ax=axesbi[0,0],xlim=[0,1])
data.groupby('SibSp')['Survived'].mean().plot(kind='barh',ax=axesbi[0,1],xlim=[0,1])
data.groupby('Parch')['Survived'].mean().plot(kind='barh',ax=axesbi[0,2],xlim=[0,1])
data.groupby('Sex')['Survived'].mean().plot(kind='barh',ax=axesbi[0,3],xlim=[0,1])
data.groupby('Embarked')['Survived'].mean().plot(kind='barh',ax=axesbi[1,0],xlim=[0,1])
sns.boxplot(x="Survived", y="Age", data=data,ax=axesbi[1,1])
sns.boxplot(x="Survived", y="Fare", data=data,ax=axesbi[1,2])
```
