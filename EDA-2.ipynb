## Isolation Forest

```python
from sklearn.ensemble import IsolationForest
```

```python
data = pd.read_csv('/home/aman/Downloads/Learning/Data science/Decision tree/iris.csv', index_col=0)

data_encoded=pd.get_dummies(data)
```

```python
# training the models
clf= IsolationForest(random_state=10,contamination=.01)
clf.fit(data_encoded)
```

```python
#Prediction
y_predict_outliers = clf.predict(data_encoded)
```

```python
y_predict_outliers
```
